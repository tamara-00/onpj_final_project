```python
prompts = [

"Write a description of an autumn morning in a melancholic style.",

"Create a short text portraying an autumn morning with a melancholic atmosphere.",

"Describe an autumn morning using a sad nostalgic literary tone.",

"Write a poetic paragraph about an autumn morning filled with melancholy.",

"Write a literary description of an autumn morning using metaphors and sensory imagery."

]
```

```python
from huggingface_hub import login

login()
```

```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
results = []
for prompt in prompts:

    for i in range(5):

        output = generator(
            prompt,
            max_new_tokens=120,
            temperature=0.7
        )[0]["generated_text"]

        results.append(output)
```

```python
print(results[0])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
embed_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings = embed_model.encode(results)
```

```python
similarity_matrix = cosine_similarity(embeddings)
```

```python
print(similarity_matrix)
```

```python
import numpy as np
```

```python
upper_triangle = similarity_matrix[np.triu_indices(len(similarity_matrix), k=1)]
```

```python
avg_cosine_similarity = np.mean(upper_triangle)
```

```python
print("Average cosine similarity:", avg_cosine_similarity)
```

```python
sentiment = pipeline("sentiment-analysis")
```

```python
scores = []

for text in results:

    result = sentiment(text[:512])[0]

    score = result["score"]

    if result["label"] == "NEGATIVE":
        score = -score

    scores.append(score)

print(scores)
```

```python
import numpy as np
```

```python
sentiment_variation = np.std(scores)
```

```python
print("Sentiment variation:", sentiment_variation)
```

```python
style_similarity = np.mean(upper_triangle)
```

```python
print("Style similarity:", style_similarity)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
def perplexity(text):

    encodings = tokenizer(text, return_tensors="pt")

    max_length = model.config.n_positions

    input_ids = encodings.input_ids[:, :max_length]

    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)

    loss = outputs.loss

    return torch.exp(loss).item()
```

```python
ppl_scores = []

for text in results:

    ppl_scores.append(perplexity(text))

print(ppl_scores)

ppl_variation = np.std(ppl_scores)

print("Perplexity variation:", ppl_variation)
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
labels = [
"melancholic poetry",
"literary prose",
"narrative fiction",
"neutral description"
]
```

```python
predicted_styles = []

for text in results:

    result = classifier(text[:512], labels)

    predicted_styles.append(result["labels"][0])

print(predicted_styles)
```

```python
from collections import Counter
```

```python
style_counts = Counter(predicted_styles)
```

```python
most_common = style_counts.most_common(1)[0][1]
```

```python
style_consistency = most_common / len(predicted_styles)
```

```python
print("Style consistency:", style_consistency)
```

```python
metrics_results = {
"cosine_similarity": avg_cosine_similarity,
"sentiment_variation": sentiment_variation,
"perplexity_variation": ppl_variation,
"style_consistency": style_consistency
}
```

```python
df_metrics = pd.DataFrame(
    list(metrics_results.items()),
    columns=["Metric", "Value"]
)
```
